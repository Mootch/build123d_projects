```python
from jupyter_cadquery import *
from build123d import *
from math import sqrt

set_defaults(orbit_control=True, axes=True, axes0=True, reset_camera=Camera.KEEP)

cv = open_viewer("SideCar", anchor="right", theme="light")
```

```python
string1 = "BOSTON"
string2 = "BRUINS"
fnsize = 20
scaled = 30
```

```python
def merge(face1, face2, amount=100.0):
    with BuildPart() as merge_fn:
        extrude(face1, amount=-amount),
        extrude(face2, amount=amount, mode=Mode.INTERSECT)
    return merge_fn.part
```

```python
with BuildPart() as sign:
    with BuildSketch(Plane.XZ.rotated((0,0,45))) as word1:
        Text(string1, font_size=fnsize, font_style=FontStyle.BOLD, align=(Align.MIN,Align.MIN))
    with BuildSketch(Plane.YZ.rotated((0,0,45))) as word2:
        Text(string2, font_size=fnsize, font_style=FontStyle.BOLD, align=(Align.MIN,Align.MIN))
    for i in range(len(string1)):
        add(merge(word1.sketch.faces().sort_by(Axis.X)[i], word2.sketch.faces().sort_by(Axis.Y)[i], max(word2.max_dimension,word1.max_dimension)))
    with BuildSketch() as base:
        with Locations((0, sign.part.bounding_box().size.Y/2)):
            RectangleRounded(sign.part.bounding_box().size.X+fnsize/2, sign.part.bounding_box().size.Y+fnsize, sign.part.bounding_box().size.X/2)
    extrude(base.sketch, amount=-fnsize/40, both=True)
    scale(sign.part, scaled/sign.part.bounding_box().size.Z)
```
